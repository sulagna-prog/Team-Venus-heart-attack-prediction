```python
#training the model #machine learning
```

```python
X = df.iloc[:,0:15]
y = df.iloc[:,15:16]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=21)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_pred = logreg.predict(X_test)
```

```python
score = logreg.score(X_test, y_test)
print("Prediction score is:",score)
```

```python
pred1=logreg.predict([[1,43,4,1,5,1,0,1,250,140,90,26,80,70,1]])
print("Predicttion 1:",pred1)
```

```python
pred2=logreg.predict([[0,20,2,0,4,0,0,1,220,100,70,24.6,60,50,1]])
print("Predicttion 2:",pred2)
```

```python
pred3=logreg.predict([[1,15,4,0,0,0,0,0,100,120,80,22,74,50,0]])
print("Predicttion 3:",pred3)
```

```python
#confusion matriix and classification report
```

```python
from sklearn.metrics import confusion_matrix, classification_report 
cm = confusion_matrix(y_test, y_pred) 
print("Confusion Matrix is:\n",cm)
```

```python

print("Classification Report is:\n\n",classification_report(y_test,y_pred))
```

```python
conf_matrix = pd.DataFrame(data = cm,  
                           columns = ['Predicted:0', 'Predicted:1'],  
                           index =['Actual:0', 'Actual:1']) 
plt.figure(figsize = (10, 6)) 
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Reds", linecolor="Blue", linewidths=1.5) 
plt.show()
```
